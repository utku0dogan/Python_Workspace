```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

matplotlib'de numpy üzerinden cok islem yapacagımızdan numpy array'e cevirdik

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g") #plt.plot(x,y) #g green 
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilo-yas")
plt.show()
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 **3
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b*-") #*-
plt.show()
```

```python
plt.plot(numpyDizisi2,numpyDizisi1,"b--") #b-- ile 
plt.show()
```

**aynı anda grafik cizmek**

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r")
plt.subplot(1,2,2) # 1 sıra olacak, 2 kolon olacak, su an 2. grafiği ciziyorum
plt.plot(numpyDizisi2,numpyDizisi1,"b--")
plt.show()
```

## grafik ozellestirme 

```python
figure2= plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7]) #bu verdiklerimizin ne ise yaradıgına bakalım
eksen2= figure2.add_axes([0.3,0.3,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("küçük Grafik Başlık")

plt.show()
```
